```python
import pandas as pd
mta = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_170617.txt")
mta.columns = mta.columns.str.strip()
```

```python
def morning_count(station_number, CA_number):
    station = mta[(mta['UNIT'] == station_number) & (mta['C/A'] == CA_number)]
    station = station.reset_index(drop=True)
    morning_dict = {}
    for i in range(len(station.index)):
        from datetime import datetime
        dates = datetime.strptime(station.DATE[i],"%m/%d/%Y")
        mytuple = (dates.strftime("%a"),station.SCP[i])
        if mytuple not in morning_dict.keys():
            morning_dict[mytuple] = [(station.TIME[i],station.EXITS[i])]
        else:
            morning_dict[mytuple].append((station.TIME[i],station.EXITS[i]))
    #return morning_dict
    station_exits = {}
    for day_and_turnstile,time_and_exits in morning_dict.items():
        station_exits[day_and_turnstile] = []
        time_and_exits.sort()
        old = 0
        for time_and_exit in time_and_exits:
            new = time_and_exit[1]
            time = datetime.strptime(time_and_exit[0], "%H:%M:%S")
            if time.strftime("%H") == "06" or time.strftime("%H") == "07" or time.strftime("%H") == "08" or time.strftime("%H") == "09":
                station_exits[day_and_turnstile].append(int(new-old))
            old = new
    #return station_exits
    morning_list = []
    for day_and_turnstile,exits in station_exits.items():
        count = 0
        for exit in exits:
            count = int(count) + int(exit)
        morning_list.append((day_and_turnstile[0], day_and_turnstile[1], count, station_number, CA_number))
    morning_list.sort()
    return morning_list
```

```python
def evening_count(station_number, CA_number):
    station = mta[(mta['UNIT'] == station_number) & (mta['C/A'] == CA_number)]
    station = station.reset_index(drop=True)
    evening_dict = {}
    for i in range(len(station.index)):
        from datetime import datetime
        dates = datetime.strptime(station.DATE[i],"%m/%d/%Y")
        mytuple = (dates.strftime("%a"),station.SCP[i])
        if mytuple not in evening_dict.keys():
            evening_dict[mytuple] = [(station.TIME[i],station.ENTRIES[i])]
        else:
            evening_dict[mytuple].append((station.TIME[i],station.ENTRIES[i]))
    #return morning_dict
    station_entries = {}
    for day_and_turnstile,time_and_entries in evening_dict.items():
        station_entries[day_and_turnstile] = []
        time_and_entries.sort()
        old = 0
        for time_and_entry in time_and_entries:
            new = time_and_entry[1]
            time = datetime.strptime(time_and_entry[0], "%H:%M:%S")
            if time.strftime("%H") == "15" or time.strftime("%H") == "16" or time.strftime("%H") == "17" or time.strftime("%H") == "18" or time.strftime("%H") == "19":
                station_entries[day_and_turnstile].append(int(new-old))
            old = new
    #return station_exits
    evening_list = []
    for day_and_turnstile,entries in station_entries.items():
        count = 0
        for entry in entries:
            count = int(count) + int(entry)
        evening_list.append((day_and_turnstile[0], day_and_turnstile[1], count, station_number, CA_number))
    evening_list.sort()
    return evening_list
```
